```python
#importing the required packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns' , 500)
```

```python
df = pd.read_csv(r'D:\Machine_Learning\Telco_churn\Data\raw\WA_Fn-UseC_-Telco-Customer-Churn (1).csv')
df.head()
```

```python
df.shape
```

```python
#check for nulls
df.isnull().sum()
```

```python
#check dtypes for each column
df.dtypes
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
#converts target column into binary values(0= not churn,1= churn)
df['Churn'] = (df['Churn']== 'Yes').astype(int)
df.head(6)
```

```python
df = df.drop(columns= 'customerID')
df
```

```python
df.columns
```

```python
#convert categorical data to numeric
df_dummies= pd.get_dummies(df, dtype=int)
df_dummies.head()
```

```python
df_dummies.isnull().sum()
```

```python
df_cleaned = df_dummies.dropna()
```
